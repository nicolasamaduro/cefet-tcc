```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn import preprocessing
from sklearn.neural_network import MLPClassifier


pd.set_option('display.max_columns', 100)
colunasPrincipais = ['atletas_pontos_num','atletas_variacao_num','atletas_media_num','local','resultado']
```

```python
rodadas = pd.read_csv('../data/dados_limpos.csv')
```

```python
def transformacaoVisitante(rodadas): 
    for index, rodada in rodadas.iterrows():
        if(rodada['local'] =='away'):
            rodadas.at[index,'atletas_variacao_num'] = - rodada['atletas_variacao_num']
            rodadas.at[index,'atletas_media_num'] = - rodada['atletas_media_num'] 
            rodadas.at[index,'atletas_pontos_num'] = - rodada['atletas_pontos_num']
    return rodadas
            
def removerJogadoresQueNaoAtuaram(rodadas):
    return rodadas.query('participou == True')
    
def removerColunasDesnecessarias(rodadas):
    return rodadas[colunasPrincipais]

def categorizarCasaVisitante(rodadas):
    def transforma(local):
        if (local=='home'):
            return 1
        else:
            return -1    
    rodadas["local"] = rodadas.apply(lambda jogador: transforma(jogador['local']), axis=1)
    return rodadas

def fazNormalizacao(rodadas):
    rodadas_normalizadas = preprocessing.MaxAbsScaler().fit_transform(rodadas)
    return pd.DataFrame(rodadas_normalizadas, columns = rodadas.columns)
    
def preprocessarDados(rodadas, transformarVisitante, normalizar, apenasParticipantes,categorizarLocal, removerColunas):
    if (apenasParticipantes):
        rodadas = removerJogadoresQueNaoAtuaram(rodadas)
    if (removerColunas):
        rodadas = removerColunasDesnecessarias(rodadas)
    if (transformarVisitante):
        rodadas = transformacaoVisitante(rodadas)
    if (categorizarLocal):
        rodadas = categorizarCasaVisitante(rodadas)
    if (normalizar):
        rodadas =fazNormalizacao(rodadas)
    return rodadas
```

```python
def realizaSorteioDosDados():
    loteria = np.random.random_sample(size=rodadas.cement.count())
    treino = [r < 0.70 for r in loteria]
    teste  = [0.70 <= r for r in loteria]
    return treino,teste

def separaCarasteristicasEClasse(rodadas):
    colunasPrincipais.pop()
    caracteristicas = rodadas.loc[:,colunasPrincipais ]
    classe = rodadas['resultado']
    return caracteristicas, classe
```

```python
def classificadorMLP(caracteristicasTreino, classeTreino, caracteristicasTeste, classeTeste):
    activationFunc ='tanh'
    clf = MLPClassifier(solver='lbfgs',activation= activationFunc, alpha=1e-5,
                    hidden_layer_sizes=(5, 5), random_state=1)
    clf.fit(caracteristicasTreino, classeTreino)
    
    clf.predict(caracteristicasTeste)
    print (round(clf.score(caracteristicasTreino,classeTreino), 4))
    print (round(clf.score(caracteristicasTeste,classeTeste), 4))
```

```python
transformarVisitante = True
normalizar = True
apenasParticipantes = True
categorizarLocal = True
removerColunas = True

preprocessar = True
if (preprocessar):
    rodadas = preprocessarDados(rodadas, transformarVisitante, normalizar, apenasParticipantes,categorizarLocal, removerColunas)

treino, teste = realizaSorteioDosDados()
caracteristicasTreino, classeTreino = separaCarasteristicasEClasse(rodadas)
caracteristicasTeste, classeTeste = separaCarasteristicasEClasse(rodadas)
classificadorMLP(treino,teste)


```
